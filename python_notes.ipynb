## args and kwargs

```python
# (*args) arguments and (**kwargs) keyword arguments
def myfunc(a, b, c=0, d=0, e=0):
    # returns 5% of the sum of a and b
    return sum((a, b, c, d, e)) * 0.05

myfunc(40,60)

# instead of doing this, we can use *args. an arbitrary number of parameters
# *args treats the parameter as a tuple of the parameters inserted
# args in *args is an arbitrary word. it can be any word as long as it has an asterisk before it
def myfunc2(*args):
     for item in args:
            print(item)

myfunc2(1, 2, 3, 4, 5)

# **kwargs are used when you want the arguments to be in dictionary format
def myfunc(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:
        print('My fruit of choice is {}'.format(kwargs['fruit']))
    else:
        print('No fruit here')
        
myfunc(fruit = 'apple', veggie = 'lettuce')

# accepting both in a function. when using both, you have to follow the order of args and kwargs parameters
def myfunc(*args, **kwargs):
    print(args)
    print(kwargs)
    print('I would like {} {}s'.format(args[0], kwargs['fruit']))
    
myfunc(10, 20, 30, fruit = 'apple', veggie = 'lettuce', animal = 'dog')
```

## Inheritance

```python
# inheritance is creating classes with classes you already created
class Animal():
    def __init__(self):
        print("Animal created")
        
    def who_am_i(self):
        print("I am an animal")
        
    def eat(self):
        print("I eat bananas")
```

```python
# deriving from the previous class
class Dog(Animal):
    
    def __init__(self):
        Animal.__init__(self)
        print("Dog created")
        
    # overwrite a previous method
    def who_am_i(self):
        print("I am a dog")
        
    def bark(self):
        print("WOOF!")
```

```python
mydog = Dog()
```

```python
# mydog now has the same methods as animal
mydog.bark()
```

```python
# creating the decorator
def new_decorator(original_func):
    
    def wrap_func():
        
        print('Some extra code, before the original function.')
        
        original_func()
        
        print("Some extra code, after the original function.")
        
    return wrap_func
```

```python
decorated_func = new_decorator(func_needs_decorator)
```

## Generators

```python
# generator functions are iterables that don't get every value at once. the iterations can be spread out to use time
# instead of creating a list with the results and then printing the list, we don't save the result
# memory efficient. no list in memory

def create_cubes(n):
    for x in range(n):
        yield x**3
```

```python
for x in create_cubes(10):
    print(x)
```

```python
# to get the list result you cast as a list
list(create_cubes(10))
```

```python
# another example

def gen_fibon(n):
    
    a = 1
    b = 2
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
for number in gen_fibon(10):
    print(number)
```

```python
# another example

def simple_gen():
    for x in range(3):
        yield x
```

```python
for number in simple_gen():
    print(number)
```

### Web scraping book example

```python
# GOAL: get the title of every book with a two-star rating
import requests
import bs4
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
base_url.format('20')
```

```python
res = requests.get(base_url.format(1))
```

```python
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
soup
```

```python
products = soup.select(".product_pod")
```

```python
example = products[0]
```

```python
# quick and dirty way to figure out if it has a rating of 2
# turn into a string and ask if it's in it

'star-ratingt Two' in str(example)
```

```python
# more official tactic of finding the answer
# if a class call has a space, you have to put a period instead of the space

[] == example.select('.star-rating.Two')
```

```python
# getting the title from the page
example.select('a')[1]['title']
```

```python
two_star_titles = []

for n in range (1,51):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(res.text, 'lxml')
    books = soup.select('.product_pod')
    
    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            book_title = book.select('a')[1]['title']
            two_star_titles.append(book_title)
```
